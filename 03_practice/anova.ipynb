```python
import pandas as pd 
import numpy as np 

group1 = np.array([5.7, 4.2, 3.3, 4.7, 4.8, 5.2, 5.2, 4.3, 5.1, 6.3])
group2 = np.array([7.3, 7.3, 4.2, 9.6, 6.2, 9.3, 6.8, 7.0, 7.8, 10.2])

df = pd.DataFrame({
    "fertilizer": ["group1"] * 10 + ["group2"] * 10,
    "growth": list(group1) + list(group2)
})

print(df)
```

```python
grand_mean = df["growth"].mean()
group_means = df.groupby("fertilizer")["growth"].mean()
group_sizes = df.groupby("fertilizer").size()

print("Grand mean:", grand_mean)
print("\nGroup means:")
print(group_means)
```

```python
# sum of squares, total
ssto = np.sum((df["growth"] - grand_mean) ** 2)

# sum of squares, between
ssb = np.sum(group_sizes * (group_means - grand_mean) ** 2)
df = df.merge(group_means.rename("group_mean"), on="fertilizer")


ssw = np.sum((df["growth"] - df["group_mean"]) ** 2)
print(f"SSTO: {ssto} \nSSB: {ssb} \nSSW: {ssw} \nCheck SSTO = SSB+SSW: {ssto} = {ssb + ssw}")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm

model = smf.ols("growth ~ fertilizer", data=df).fit()
anova_table = anova_lm(model)

print(anova_table)
```
