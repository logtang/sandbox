# **t-Testing**

```python
# Two-Sample t-Test Example
import numpy as np
from scipy import stats

# Loading the data
group1 = np.array([5.7, 4.2, 3.3, 4.7, 4.8, 5.2, 5.2, 4.3, 5.1, 6.3])
group2 = np.array([7.3, 7.3, 4.2, 9.6, 6.2, 9.3, 6.8, 7.0, 7.8, 10.2])

# Under the hood, a unified t-statistic for both samples, along with a pooled sd, etc
t_stat, p_value = stats.ttest_ind(group1, group2)
print("t statistic:", t_stat)
print("p value:", p_value)
```
