# Linear Regression

#### **The Simple Model (Intercept and One Slope Coefficient)**

```python
import pandas as pd
import numpy as np

# Loading data (fish length in cm, fish weight in grams)
length = np.array([10, 12, 13, 15, 16, 18, 20])
weight = np.array([100, 130, 135, 160, 180, 210, 230])

df = pd.DataFrame({
    "length": length,
    "weight": weight
})

print(df)
```

```python
# Manually computing slope and intercept

x = df["length"]
y = df["weight"]

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

beta_1 = numerator / denominator
beta_0 = y_mean - beta_1 * x_mean

print(f"Beta 0, the Intercept: {beta_0} \nBeta 1, Slope Coefficient: {beta_1}")
```

```python
# Adding col for predicted weight
df["predicted_weight"] = beta_0 + beta_1 * df["length"]
print(df)
```

```python
# Plotting the model with plotly
import plotly.express as px

fig = px.scatter(df, x="length", y="weight")

fig.add_scatter(
    x=df["length"],
    y=df["predicted_weight"],
    mode="lines",
    name="Regression Line"
)

fig.show()
```

```python
# Fitting the simple model with scipy
from scipy.stats import linregress
result = linregress(length, weight)
print(f"Beta 0, Intercept: {result.intercept} \nBeta 1, Slope Coefficient: {result.slope}")
```

```python
intercept = result.intercept; slope = result.slope
predicted = intercept + slope * length
df["predicted_weight"] = intercept + slope * df["length"]

import plotly.express as px

fig = px.scatter(df, x="length", y="weight")
fig.add_scatter(
    x=df["length"],
    y=df["predicted_weight"],
    mode="lines",
    name="Regression Line"
)

fig.show()
```
